```python
def Auto_petal_plot(v,labels,selfcolor=None,labelsize=20,title=' ',ifsave='no',savename=None):
    import numpy as np
    import matplotlib.pyplot as plt

    plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams['axes.unicode_minus']=False

    if selfcolor is not None and isinstance(selfcolor, str):
        try:
            selfcolor = getattr(plt.cm, selfcolor)
        except AttributeError:
            selfcolor = plt.cm.viridis
    elif selfcolor is None:
         selfcolor = plt.cm.viridis

    v_array = np.array(v)
    labels_array = np.array(labels)

    if len(v_array) == 0:
        fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16), facecolor='white')
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(title, y=1.05, fontsize=labelsize)
        plt.show()
        if ifsave == 'yes' and savename:
            fig.savefig(savename, dpi=300)
        return

    sorted_indices = np.argsort(v_array)[::-1]
    labels_sorted = labels_array[sorted_indices]
    v_values_sorted = v_array[sorted_indices]

    N = len(v_values_sorted)
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)

    min_val_orig = np.min(v_values_sorted)
    max_val_orig = np.max(v_values_sorted)
    range_val_orig = max_val_orig - min_val_orig

    scaled_min_display = 0.5
    scaled_max_display = 1.5
    if N == 0: 
        heights_for_plot = np.array([])
    elif range_val_orig < 1e-9:
        heights_for_plot = np.full_like(v_values_sorted, scaled_min_display, dtype=float)
    else:
        heights_for_plot = (v_values_sorted - min_val_orig) / range_val_orig * (scaled_max_display - scaled_min_display) + scaled_min_display

    gap = 0.1
    width = (2 * np.pi / N) * (1 - gap)

    fig, ax = plt.subplots(subplot_kw={'polar': True}, figsize=(16,16),facecolor = 'white')
    ax.spines['polar'].set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    if not(isinstance(selfcolor, list)):
        colors_to_use = selfcolor(np.linspace(0.2, 0.8, N))
    else:
        colors_to_use=selfcolor
    bars = ax.bar(theta, heights_for_plot, width=width, bottom=0.0, 
                  color=colors_to_use, 
                  edgecolor='none')

    max_plot_height_val = np.max(heights_for_plot) if N > 0 else 1.0
    label_radial_offset = max_plot_height_val * 0.2 +((labelsize-20)*0.005)

    for i in range(N):
        bar_plotted_height = heights_for_plot[i]
        original_value_for_label = v_values_sorted[i]
        current_label_text = labels_sorted[i]
        
        angle_rad_for_position = theta[i]

        r_label_position = bar_plotted_height + label_radial_offset
        
        angle_deg_of_petal = np.degrees(angle_rad_for_position)
        
        text_rotation_tangential = angle_deg_of_petal + 90.0
        final_text_rotation = text_rotation_tangential
        normalized_angle = final_text_rotation % 360
        if normalized_angle > 90 and normalized_angle < 270:
            final_text_rotation += 180.0
        
        ax.text(angle_rad_for_position, r_label_position, f'{current_label_text}: {original_value_for_label:.2f}',
                ha='center', va='center', fontsize=labelsize,
                rotation=final_text_rotation, rotation_mode='anchor')

    ax.set_theta_zero_location('N')
    ax.set_title(title,y = 1.2,fontsize = labelsize)
    plt.show()
    
    if ifsave == 'yes' and savename is not None:
        fig.savefig(savename, dpi=300)
```

```python
def Auto_pie_plot(v, labels, selfcolor=None, labelsize=20, title=' ', ifsave='no', savename=None):
    import numpy as np
    import matplotlib.pyplot as plt
    import matplotlib.colors
    import math
    plt.rcParams['font.sans-serif']=['SimHei']
    plt.rcParams['axes.unicode_minus']=False
    
    processed_color_input = selfcolor 

    if processed_color_input is not None and isinstance(processed_color_input, str):
        try:
            processed_color_input = getattr(plt.cm, processed_color_input)
        except AttributeError:
            processed_color_input = plt.cm.viridis
    elif processed_color_input is None:
         processed_color_input = plt.cm.viridis

    v_array = np.array(v)
    labels_array = np.array(labels)
    
    fig, ax = plt.subplots(figsize=(16,16), facecolor='white')

    if len(v_array) == 0:
        ax.axis('off') 
    else:
        valid_indices = v_array > 0
        v_positive = v_array[valid_indices]
        labels_positive = labels_array[valid_indices]

        if len(v_positive) == 0:
            ax.axis('off')
        else:
            initial_sorted_indices = np.argsort(v_positive)[::-1] 
            v_initial_sorted = v_positive[initial_sorted_indices]
            labels_initial_sorted = labels_positive[initial_sorted_indices]
            
            N = len(v_initial_sorted)
            
            colors_to_use = None
            if isinstance(processed_color_input, matplotlib.colors.Colormap):
                colors_to_use = processed_color_input(np.linspace(0.2, 0.8, N))
            elif isinstance(processed_color_input, list):
                colors_to_use = [processed_color_input[i % len(processed_color_input)] for i in range(N)] if len(processed_color_input) < N else processed_color_input[:N]
            else: 
                colors_to_use = plt.cm.viridis(np.linspace(0.2, 0.8, N))

            total_value_original = np.sum(v_initial_sorted)
            original_percentages = (v_initial_sorted / total_value_original) * 100.0 if total_value_original > 0 else np.zeros_like(v_initial_sorted)
            
            threshold_percentage = 5.0
            exponent = 0.6 
            v_for_pie_display = np.power(v_initial_sorted.astype(float) + 1e-9, exponent)

            pie_display_labels_text = [f'{label}: {value:.2f}' for label, value in zip(labels_initial_sorted, v_initial_sorted)]

            wedges, texts_pie_labels, texts_autopct_objects = ax.pie(
                v_for_pie_display,
                labels=pie_display_labels_text,
                colors=colors_to_use,
                autopct=lambda p: '', 
                startangle=90,      
                pctdistance=0.80,   
                labeldistance=1.05, 
                wedgeprops=dict(edgecolor='w')
            )

            plt.setp(texts_pie_labels, fontsize=int(labelsize * 0.7), wrap=True) 
            plt.setp(texts_autopct_objects, fontsize=int(labelsize * 0.6), color="white", weight="bold")

            small_slices_info_for_external = []
            sp_label_obj_on_pie = None
            cr_label_obj_on_pie = None

            for i, wedge_obj in enumerate(wedges):
                if texts_autopct_objects and i < len(texts_autopct_objects):
                     texts_autopct_objects[i].set_text(f'{original_percentages[i]:.2f}%')
                
                current_label_name_on_pie = labels_initial_sorted[i]
                is_large_slice_on_pie = not (original_percentages[i] < threshold_percentage and total_value_original > 0)

                if is_large_slice_on_pie and texts_pie_labels[i].get_visible():
                    if current_label_name_on_pie == "SP":
                        sp_label_obj_on_pie = texts_pie_labels[i]
                    elif current_label_name_on_pie == "C_R":
                        cr_label_obj_on_pie = texts_pie_labels[i]

                if original_percentages[i] < threshold_percentage and total_value_original > 0:
                    texts_pie_labels[i].set_visible(False)
                    if texts_autopct_objects and i < len(texts_autopct_objects):
                        texts_autopct_objects[i].set_visible(False)
                    
                    external_text = f'{labels_initial_sorted[i]}: {v_initial_sorted[i]:.2f}'
                    angle_deg_wedge_center = (wedge_obj.theta1 + wedge_obj.theta2) / 2.0
                    angle_deg_wedge_center = angle_deg_wedge_center % 360 

                    small_slices_info_for_external.append({
                        'wedge_obj': wedge_obj,
                        'text': external_text,
                        'original_value': v_initial_sorted[i], 
                        'angle_deg_center': angle_deg_wedge_center,
                        'label_name': labels_initial_sorted[i] 
                    })
            
            small_slices_info_for_external.sort(key=lambda x: x['original_value'])

            pie_radius = wedges[0].r if wedges else 1.0
            
            num_small_labels = len(small_slices_info_for_external)
            sp_x_max_for_ep = -pie_radius * 2.0 
            cr_x_min_for_t2m = pie_radius * 2.0  
            fig.canvas.draw() 

            if sp_label_obj_on_pie:
                sp_pos = sp_label_obj_on_pie.get_position()
                if sp_pos[0] < 0: 
                    sp_x_max_for_ep = sp_pos[0] + pie_radius * 0.1 
                else: 
                    sp_x_max_for_ep = sp_pos[0] - pie_radius * 0.3


            if cr_label_obj_on_pie:
                cr_pos = cr_label_obj_on_pie.get_position()
                if cr_pos[0] > 0:
                    cr_x_min_for_t2m = cr_pos[0] - pie_radius * 0.1 
                else: #
                    cr_x_min_for_t2m = cr_pos[0] + pie_radius * 0.3

            y_text_level = pie_radius * 1.20 
            title_y_position = 1.05 
            if num_small_labels > 0:
                title_y_position = 1.1 

            x_coords_final_for_small = np.zeros(num_small_labels)

            if num_small_labels > 0:
                max_x_deviation_of_text = pie_radius * 0.7 

                if num_small_labels == 1:
                    x_coords_final_for_small[0] = 0.0
                    info_item = small_slices_info_for_external[0]
                    if info_item['label_name'] == "E_P":
                        x_coords_final_for_small[0] = min(0.0, sp_x_max_for_ep - 0.03 * pie_radius)
                    elif info_item['label_name'] == "T2M":
                        x_coords_final_for_small[0] = max(0.0, cr_x_min_for_t2m + 0.03 * pie_radius)
                else:
                    tentative_x_positions = np.linspace(-max_x_deviation_of_text, 
                                                         max_x_deviation_of_text, 
                                                         num_small_labels)
                    
                    for k_idx, info_item in enumerate(small_slices_info_for_external):
                        current_x = tentative_x_positions[k_idx]
                        
                        if info_item['label_name'] == "E_P":
                            current_x = min(current_x, sp_x_max_for_ep - 0.03 * pie_radius)
                        elif info_item['label_name'] == "T2M":
                            current_x = max(current_x, cr_x_min_for_t2m + 0.03 * pie_radius)
                        
                        x_coords_final_for_small[k_idx] = current_x
                    
                    ep_constrained_x, t2m_constrained_x = None, None
                    ep_original_k_idx, t2m_original_k_idx = -1, -1

                    for k_idx, info_item in enumerate(small_slices_info_for_external):
                        if info_item['label_name'] == "E_P":
                            ep_constrained_x = x_coords_final_for_small[k_idx]
                            ep_original_k_idx = k_idx
                        elif info_item['label_name'] == "T2M":
                            t2m_constrained_x = x_coords_final_for_small[k_idx]
                            t2m_original_k_idx = k_idx
                    
                    if ep_original_k_idx != -1 and t2m_original_k_idx != -1 and \
                       ep_original_k_idx < t2m_original_k_idx and \
                       ep_constrained_x > t2m_constrained_x:

                        mid_point = (ep_constrained_x + t2m_constrained_x) / 2.0
                        gap = pie_radius * 0.05
                        x_coords_final_for_small[ep_original_k_idx] = mid_point - gap
                        x_coords_final_for_small[t2m_original_k_idx] = mid_point + gap

            for i, info in enumerate(small_slices_info_for_external):
                current_text_x = x_coords_final_for_small[i]
                current_text_y = y_text_level
                
                wedge = info['wedge_obj']
                target_angle_on_wedge_deg = info['angle_deg_center']
                target_angle_rad = np.deg2rad(target_angle_on_wedge_deg)
                x_wedge_conn = pie_radius * np.cos(target_angle_rad)
                y_wedge_conn = pie_radius * np.sin(target_angle_rad)

                ax.annotate(info['text'],
                            xy=(x_wedge_conn, y_wedge_conn),
                            xytext=(current_text_x, current_text_y),
                            textcoords='data',
                            fontsize=int(labelsize * 0.6),
                            arrowprops=dict(arrowstyle="-",
                                            connectionstyle="arc3,rad=0",
                                            color='black',
                                            linestyle='--',
                                            shrinkA=5, 
                                            shrinkB=10),
                            bbox=dict(boxstyle="round,pad=0.25", fc="white", ec="gray", lw=0.5, alpha=0.9),
                            horizontalalignment='center', 
                            verticalalignment='center',
                            wrap=False)
            ax.axis('equal')  
    
    ax.set_title(title, y=title_y_position, fontsize=labelsize) 
    plt.show()
    
    if ifsave == 'yes' and savename is not None:
        fig.savefig(savename, dpi=300, bbox_inches='tight')
```
